## List Comprehenssion

```python
p = [1,4,7,34,65,34]
x= [i**2 for i in p if i%2==1]
print(x)
```

```python
p = [1,4,7,34,65,34,15]
y = [i for i in p if i%3==0 and i%5==0]
print(y)
```

```python
p = [1,4,7,34,65,34,15]
z = [(i**2,i**3) for i in p if i%2==1]
print(z)
```

```python
t = ("Virat","dhoni","sachin","Arjun")
t = tuple([len(i) for i in t])
print(t)
```

```python
l = []
for i in range(3):
    s = []
    for j in range(3):
        s.append(0)
    l.append(s)
print(l)

```

```python
s = [[0 for j in range(3)] for i in range(3) ]
s
```

```python
p = [1,4,7,34,65,34,15]
l = [i**2 if i%2==0  else i if i%3==0 else i**3 for i in p]
print(l)
```

```python
l = [[[i for i in range(1,4)] for j in range(1)] for k in range(1)]
print(l)
```

```python
w=[12,23,45,56,67,67,78]
l=[ i**2 if i%4==0 else  i**3 if i%5==0 else i*i if i%3==0 else i for i in w]
print(l)
```

```python
w=[12,23,45,56,67,67,78]
l=[]
for i in w:
    if i%4==0:
        l.append(i**2)
    elif i%5==0:
        l.append(i**3)
    elif i%3==0:
        l.append(i*i)
    else:
        l.append(i)
print(l)
```
