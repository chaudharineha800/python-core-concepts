# Exception handling 
###  in Python is used to handle runtime errors (exceptions) so that the program does not crash and can continue executing normally.
### Exception handling in Python is the process of handling runtime errors using try, except, else, and finally blocks to prevent program termination.
## 📌 What is an Exception?

### An exception is an error that occurs during program execution.

Examples:

Division by zero

File not found

Invalid input

Index out of range

## 🎯 Why Exception Handling is Important?

✅ Prevents program crash
✅ Handles errors gracefully
✅ Improves user experience
✅ Helps debugging

## ✅ Advantages

✔ Prevents abnormal termination
✔ Maintains normal program flow
✔ Helps in debugging
✔ Provides meaningful messages

| Exception           | Meaning                |

| `ZeroDivisionError` | Division by zero       |
| `ValueError`        | Invalid value          |
| `TypeError`         | Wrong data type        |
| `IndexError`        | List index error       |

### Try–Except Block (Basic Syntax)
### try:
###     risky code
### except:
###    error handling code

```python
try:
    a = int(input("Enter number: "))
    b = 10 / a
    print(b)
except:
    print("Error occurred")
```

```python
# Handling Specific Exceptions
try:
    a = int(input("Enter number: "))
    print(10 / a)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input")
```

```python
# Multiple Exceptions in One Line
try:
    a = int(input("Enter number: "))
    print(10 / a)
except (ZeroDivisionError, ValueError):
    print("Error occurred")
```

```python
# Else Block
#else runs only if no exception occurs.

try:
    a = 10 / 2
except:
    print("Error")
else:
    print("Success:", a)
```

```python
# Finally Block
#finally always executes whether error occurs or not.

try:
    f = open("data.txt")
except:
    print("File not found")
finally:
    print("Program finished")

# Common use: closing files, releasing resources.
```

### Full Structure
### try:
    #code
### except ExceptionType:
    #handle error
### else:
    #runs if no error
### finally:
    #always runs

```python
# Raising Exceptions (raise)

# You can create your own exceptions.

age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

```python
# User Defined Exception
class MyError(Exception):
    pass

try:
    raise MyError("Custom error")
except MyError as e:
    print(e)
```

```python
try:
    num = int(input("Enter number: "))
    result = 100 / num
    print("Result:", result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Please enter valid number")

finally:
    print("Execution completed")
```

### ❌ Errors That Cannot Be Handled
## Syntax Errors

These occur before execution, so try-except cannot catch them.

Example:

if True
    print("Hello")

SyntaxError happens before program runs.

🎯 Simple Rule

### Exception handling can handle runtime errors but not compile-time (syntax) errors.

📌 Exam Answer (Short)

### Exception handling can handle runtime errors such as ZeroDivisionError, ValueError, TypeError, IndexError, FileNotFoundError, etc., but it cannot handle syntax errors.

### Syntax errors cannot be handled using normal exception handling because they occur before program execution. They must be corrected manually. However, they can be caught using exec() or eval() when executing dynamic code.

## ✅ 1. Fix the Syntax Error (Main Method)

Carefully check:

Missing : (colon)

Incorrect indentation

Missing brackets

Wrong keywords spelling

❌ Wrong:
if True
    print("Hello")

✅ Correct:
if True:
    print("Hello")

## ✅ 2. Using try-except with exec() or eval() (Advanced)

If you are running dynamic code (user input code), you can catch syntax errors.

```python
code = "if True print('Hello')"

try:
    exec(code)
except SyntaxError:
    print("Syntax error detected")

# Here SyntaxError can be caught because the code is executed dynamically.
```

# 🔹 eval() Function

### eval() is used to evaluate an expression and return the result.

### 👉 It works only with single expressions (not full programs).

### Syntax
### eval(expression)

```python
x = 10
result = eval("x + 5")
print(result)
```

#### ✅ Returns value
#### ❌ Cannot execute statements like loops or assignments

# 🔹 exec() Function

### exec() is used to execute a block of code (multiple statements).

### 👉 It does not return a value.

### Syntax
### exec(code)

```python
code = """
a = 10
b = 20
print(a + b)
"""

exec(code)
```

| Feature            | eval()              | exec()             |
| ------------------ | ------------------- | ------------------ |
| Purpose            | Evaluate expression | Execute statements |
| Returns value      | ✅ Yes               | ❌ No               |
| Multiple lines     | ❌ No                | ✅ Yes              |
| Assignment allowed | ❌ No                | ✅ Yes              |
| Loops allowed      | ❌ No                | ✅ Yes              |


🔹 Handling Syntax Error with exec()
code = "if True print('Hello')"

try:
    exec(code)
except SyntaxError:
    print("Syntax Error detected")
⚠️ Security Warning

Never use eval() or exec() with untrusted user input.

Example dangerous code:

eval("__import__('os').system('delete files')")

It can run harmful commands.

### eval() evaluates a Python expression and returns the result, while exec() executes a block of Python code without returning a value.
