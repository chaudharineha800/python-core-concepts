# Inheritance in Python (Easy Descriptive Explanation)

Inheritance is one of the most important concepts of Object-Oriented Programming (OOP). Inheritance allows one class to acquire the properties and methods of another class. In simple words, inheritance means creating a new class from an existing class so that we can reuse the existing code instead of writing it again.

The class whose properties are inherited is called the parent class (base class or superclass), and the class that inherits those properties is called the child class (derived class or subclass).

Inheritance helps programmers to reduce code duplication, improve code reusability, and make programs easier to maintain.

For example, if we create a class called Animal that contains a method like eat(), then another class Dog can inherit from Animal and automatically use the eat() method without rewriting it.

Syntax of Inheritance

In Python, inheritance is implemented by passing the parent class name inside the parentheses while defining the child class.

# Example:

# class Parent:
#     pass

# class Child(Parent):
#     pass


Here, Child class inherits properties from Parent class.

Example of Inheritance
class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    def bark(self):
        print("Dog is barking")

d = Dog()
d.eat()
d.bark()


In this example, Dog class inherits the eat() method from Animal class. So the Dog object can use both eat() and bark() methods.

# Types of Inheritance in Python

Python supports five types of inheritance.

1. Single Inheritance

In single inheritance, one child class inherits from one parent class.

Example: Father → Son

2. Multiple Inheritance

In multiple inheritance, one child class inherits from more than one parent class.

Example: Child inherits from both Mother and Father.

3. Multilevel Inheritance

In multilevel inheritance, inheritance happens in multiple levels.

Example: Grandfather → Father → Son

4. Hierarchical Inheritance

In hierarchical inheritance, multiple child classes inherit from one parent class.

Example: One parent with many children.

5. Hybrid Inheritance

Hybrid inheritance is a combination of two or more types of inheritance.

Method Overriding in Inheritance

Sometimes the child class wants to change the behavior of a parent method. This is called method overriding.

Example:

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")


Here the Dog class overrides the sound() method.

Method Resolution Order (MRO)

When multiple inheritance is used, Python needs to decide which method should be executed first. This order is called Method Resolution Order (MRO).

Python automatically determines the order based on class hierarchy.

# What does super() return in Python?

## The super() function returns a temporary object of the parent (super) class that allows you to call its methods.

In simple words:

## 👉 super() does NOT return a value like a number or string.
## 👉 It returns a reference (proxy object) to the parent class.

That reference is then used to access parent methods or constructors.

Easy Definition

## super() returns an object that represents the parent class, so we can call parent methods using it.

Parent __init__() gets called.

## super() → reference to class A → calls show().

What Exactly super() Returns (Technical)

Technically:

## super() returns a super object (proxy object) that delegates method calls to the parent class according to the MRO (Method Resolution Order).

```python
class emp:
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal
    def get_details(self):
        print("Name: ",self.name)
        print("Salary: ",self.sal)

class mgr(emp):
    def __init__(self,name,sal,dept):
        super().__init__(name,sal)
        self.dept=dept
    def get_details(self):
        super().get_details()
        print("Department: ", self.dept)

```

# Abstraction
### Abstraction in python is a fundamenal concept in object oriented programming that focuses on hiding implementation details and showing only the essential features of an object. It allows you to reduce complexity and increase effieciency by exposing only necessary part of an object yo yhe user.
## Key Points:
### Hiding complexity: Abstraction enables you to hide the complicated underlying logic and only expose a simpler interface for interation
### Interface and Abstract Class: 
### Use Cases: 

## Key Concepts:
### 1.Abstract class:
#### An Abstract class is a class that cannot be instantiated directly. It serves as a blueprint for other classes.It may contain abstract (methods without inplementation) as well as concrete methods(methods with implementation)
### 2. Abstract Method:
#### An Abstract method is a method that is declared but contians no implementation> It is meant to be overrideen in derived class. Any classs containing an abstract method must also be declared as an abstract class
### 3. ABC Module:
#### The ABC module in Python, Which stands for Abstract Base Case, Provides tools for defining abstract classes and methods.Yoou need to import ABC and abstract methods from the abc module to create an abstract class and define abstract methods;

## 👉 Can we override concrete methods of an abstract class?

You can say:

### Yes, concrete methods of an abstract class can be overridden in the derived class, just like normal inheritance. It is optional, while abstract methods must be implemented.

Method Type	Override Required?

### We cannot create an object of an abstract class because:
### An abstract class is incomplete (it may contain abstract methods without implementation).
### Since some methods don’t have a body, Python does not know how the object should behave fully.

So object creation is restricted.

### Yes, an abstract class can have a constructor. It is used to initialize common attributes, and it gets called when an object of the derived class is created.

### ✅ Quick Summary
Feature	Abstract Class
Constructor Allowed	✅ Yes
Object Creation	❌ No (if abstract methods exist)
Constructor Called When	Child object created

```python
from abc import ABC, abstractmethod
class A(ABC):
    @abstractmethod
    def method1(self):
        pass
    def method2(self,p1):
        self.p1=p1
class B(A):
    def method2(self,p1):
        super().method2(p1)
    def method1(self):
        print(self.p1)

b= B()
b.method2(12)
b.method1()
```

#  Encapsulation
### Encapsulation is a fudamental concept in object oriented programming where the internal representation of an object is hidden from the outside world. It ensures that the data(attribute) and the method(function) that operates on the data are bundled together and access to them is restricted or controlled
### in python, encapsulation is typically achieved using private and protected access specifiers
## Access Specifiers:
### Public: Accessibile from anywhere(default)
### Protected: Ondicated by s single underscore _, intended to be used by the class and its subclass
### Private : indicated by a double underscore __, not accessible directly outside the class(name mangling in python)

```python
class A:
    def __init__(self,a,b,c):
        self.a = a#public
        self._b=b#protected 
        self.__c=c#privte
a = A(1,2,3)
print(a.a)
print(a._b)
print(a._A__c)
```

# ⭐ Important Interview Line

## Name mangling is the process of modifying variable names internally by Python to avoid naming conflicts and provide limited access to private variables.

# 📦 Real-Life Analogy

Think of it like:

Your phone contact name: Mom

But internally phone saves it as:

Contact_001_Mom


You don’t see it, but system does.

That is name mangling.

🎯 Why Name Mangling is Used

To avoid accidental access

To prevent conflicts in inheritance

To indicate variable is private/internal

Name mangling means:

# Python automatically changes the name of a variable internally to make it harder to access from outside the class.

# It is mainly used for private variables (those starting with __).

🔐 Simple Definition

# Name Mangling = Renaming a variable internally by Python

# 🔐 Why Python Uses Name Mangling

In Python, there is no strict private access modifier like Java or C++.

Instead, Python uses name mangling to:

## ✅ Avoid accidental access
## ✅ Prevent name conflicts in inheritance
## ✅ Indicate “this variable is for internal use only”

🧠 What is Name Mangling?

When you write:

self.__c = 10


Python internally converts it to:

self._ClassName__c


So in your class A, this:

self.__c


becomes:

self._A__c

🔎 Example
class A:
    def __init__(self):
        self.__x = 100

obj = A()

print(obj.__x)      # ❌ Error
print(obj._A__x)    # ✅ Works


Output:

AttributeError
100

🚀 Why Not Completely Private?

Python philosophy:

“We are all adults here.”

Meaning:

Python trusts developers.

It does not strictly block access.

It only discourages misuse.

🧬 Main Reason — Inheritance Safety

Imagine this situation:

class Parent:
    def __init__(self):
        self.__value = 10

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.__value = 20


Without name mangling → conflict happens.

With name mangling:

Parent → _Parent__value
Child  → _Child__value


Both exist safely ✅

🔥 When to Use __private?

Use double underscore when:

✔ You don’t want subclasses to override accidentally
✔ Variable is very sensitive/internal
✔ Avoid naming conflicts

## Getter and setter
### In encapsulation , we often restict direct access to attribute by making them as protected and private. Instead we provide getter and setter methods which control how the attributes are accessed or modified
### Gwtter: A method that returns the value of an attribute
### Setter: A mehtod that sets the value of an attribute, often with validation

```python
class Employee:

    def __init__(self, salary):
        self.__salary = salary   # private variable

    # Getter
    def get_salary(self):
        return self.__salary

    # Setter
    def set_salary(self, salary):
        if salary > 0:
            self.__salary = salary
        else:
            print("Invalid salary")

emp = Employee(50000)

print(emp.get_salary())   # Getter
emp.set_salary(60000)     # Setter
print(emp.get_salary())
```

# Polymorphism
### Polymorphism is one of the key principle in object-Oriented Programming. It refers to the abilitt of different object to respond to the same method or operation in different ways. Polymorphism allows objects of different types to be treated as if they were object of the same type, without the need to know their specific class.
### Lets explore method overriding (dynamic polymorphism) in python , which allows child classes to provide their own specific implementation of methods that are already defined in the parent class.
### Polymorphism is the ability of an object, function, or operator to take multiple forms depending on the context.

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

🔷 What is Polymorphism in Python (OOP)?

Polymorphism means:

One name — many forms

In simple words:

👉 The same function or method can behave differently in different situations.

📱 Real-Life Example

Think of a person:

At home → Son/Daughter

At office → Employee

With friends → Friend

Same person, different roles → Polymorphism.

🧠 Types of Polymorphism in Python

There are mainly 2 types:

1️⃣ Method Overriding (Runtime Polymorphism)

Child class changes the method of parent class.

class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()


Output:

Dog barks

2️⃣ Method Overloading (Compile-Time Concept)

Python does not support true overloading, but we can simulate it using default arguments.

class Math:

    def add(self, a, b=0, c=0):
        print(a + b + c)

m = Math()

m.add(5, 10)
m.add(5, 10, 15)


Output:

15
30

🔥 Operator Polymorphism

Same operator behaves differently.

print(5 + 5)        # addition → 10
print("Hi " + "All") # concatenation → Hi All


+ works differently → Polymorphism.

🎯 Function Polymorphism

Same function works with different data types.

len("Hello")
len([1,2,3,4])
