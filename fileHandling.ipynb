# File Handling 
### File handling is an essential aspect of programming as it allows you to interact with files on your computer You can read from,write to and manipulate files using pythons built in file handling functions.
### Python provides several built in functions to handle fileoperations such as:
#### open(): Open a file and return a file object
#### read(): Reds the content of file
#### write(): Write data to a file
#### close(): closes file after operations
#### with Statement: Auctomatically handles file closing, ensuring that resources are properly managed.

## Modes of files to open:
### Read mode('r'): Opens file for reading. The file must exist otherwise an error raised
### Write mode('w'): Opens the file for writing, if file dosent exist it creates it and write into it. It chnages whole previsous content to new content
### Append mode('a'): Opens the file for appending, if the file exists, data is added to the end of the file without trauncating it. It it dosent exists, a new file is created.
### Exclusive Creation Mode('x'): Opens file for writing but fails if the file already exists. Useful for avoiding overwriting existing file.

| Mode  | Meaning                           |
| ----- | --------------------------------- |
| `"r"` | Read (default)                    |
| `"w"` | Write (creates new or overwrites) |
| `"a"` | Append (adds data at end)         |
| `"x"` | Create file (error if exists)     |
| `"b"` | Binary mode                       |
| `"t"` | Text mode (default)               |
| `"+"` | Read and write both               |

```python
# Read entire file
f = open("k315.txt", "r")
print(f.read())
f.close()

# Read One Line
f = open("k315.txt", "r")
print(f.readline())
f.close()

# Read All Lines as List
f = open("k315.txt", "r")
print(f.readlines())
f.close()
```

```python
# Checking File Properties
f = open("k315.txt", "r")

print(f.name)      # File name
print(f.mode)      # Mode
print(f.closed)    # True or False

f.close()
```

# Pickling and Unpickling
### Pickling = Converting a Python object into a byte stream (binary format) so it can be stored in a file or transferred over a network.

👉 In simple words: Saving object permanently

Example objects:

List

Dictionary

Class objects

Tuples

Numbers

### Unpickling = Converting the byte stream (file data) back into the original Python object.

👉 In simple words: Loading object back into program

| Function                 | Purpose                 |
| ------------------------ | ----------------------- |
| `pickle.dump(obj, file)` | Save object to file     |
| `pickle.load(file)`      | Read object from file   |
| `pickle.dumps(obj)`      | Convert object to bytes |
| `pickle.loads(bytes)`    | Convert bytes to object |


## ⚠️ Important Points

✅ Works only with Python
✅ File must be opened in binary mode (rb, wb)
✅ Faster than text storage
❌ Not secure for untrusted data (can execute malicious code)

🎯 Simple Exam Definition

### Pickling is the process of converting a Python object into a byte stream to store it in a file, and unpickling is the process of converting the byte stream back into the original object.

##👍 Real-Life Uses

Saving machine learning models

Storing program state

Caching data

Sending objects over network

| Extension | Meaning           | Usage                          |
| --------- | ----------------- | ------------------------------ |
| `.dat`    | Generic data file | Educational / general programs |
| `.pkl`    | Pickle file       | Professional Python projects   |
| `.pickle` | Pickle file       | Less common                    |


```python
import pickle

data = {"name": "Neha", "age": 20, "course": "BCA"}

# open file in binary write mode
with open("student.dat", "wb") as f:
    pickle.dump(data, f)

print("Data pickled successfully")
```

```python
with  open("student.dat","r") as f:
    print(f.read())
```

```python
import pickle

with open("student.dat", "rb") as f:
    data = pickle.load(f)

print(data)
```
