#### The scraper extracts required data from HTML.

```python
import requests
from bs4 import BeautifulSoup
url="https://books.toscrape.com"
req = requests.get(url)
webContent = BeautifulSoup(req.text,"html.parser")  
print("done")
```

```python
title = webContent.find_all("h3")
title_list = [t.text  for t in title]
print(title_list)
```

```python
price = webContent.find_all("p", class_="price_color")
price_list = [ p.text for p in price ]
print(price_list)
```

```python
inStock = webContent.find_all("p",class_="instock availability")
inStock_list = [ i.text.strip() for i in inStock]
print(inStock_list)
```

```python
links = webContent.find_all("a")

links_list=[ link.get("href") for link in links]
print(links_list)
```

```python
data = [[i,j,k] for i,j,k in zip(title_list, price_list,inStock_list)]
print(data)
```

```python
import pandas as pd
df = pd.DataFrame(data,columns=["Book_Name","Price","Is_In_Stock"])
df.to_csv("Books.csv",index=False)
print("Data Saved Successfully")
```
